## D_r(real data)

```python
import numpy as np
import pandas as pd
import plotly.express as px
from flowgen import stock_5y, synthesis, regime_analysis, plot_regime_switching
```

```python
nvda = stock_5y("NVDA")
```

```python
nvda.head()
```

```python
fig = px.line(nvda.close, title='NVIDIA Closing Price')
fig.show()
```

```python
# trend component of closing price
## T = (at+b)(sin(t)) + (ct+b)

#
```

### Regime Switching Analysis on Real Data

```python
nvda_regime = regime_analysis(nvda)
```

```python
plot_regime_switching(nvda, nvda_regime)
```

## Synthetic Data Generation D_g

```python
nvda_synth = synthesis(nvda)
synth_v_real = pd.DataFrame({'Synthetic': nvda_synth['close'], 'Actual': nvda['close']})
```

```python
fig = px.line(synth_v_real, y=synth_v_real.columns,
             title='Actual vs Synthetic')
fig.show()
```

### Regime Switching Analysis on Synthetic Data

```python
nvda_synth_regime = regime_analysis(nvda_synth)
```

```python
plot_regime_switching(nvda_synth, nvda_synth_regime)
```
